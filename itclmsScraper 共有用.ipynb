```python
#関数化(完成版)
def download(filepath,url):
  os.mkdir(filepath)
  os.chdir(filepath)
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  driver = webdriver.Chrome('chromedriver',options=options)
  driver.get(url)  
  driver.find_element_by_class_name('square_button').click()
  sleep(5)
  #ログインフォームに入力
  ID='********@utac.u-tokyo.ac.jp'#ログイン時のIDに書き換え
  PASS='************'#パスワードに書き換え
  driver.find_element_by_id('userNameInput').send_keys(ID)
  driver.find_element_by_id('passwordInput').send_keys(PASS)
  driver.find_element_by_id('submitButton').click()
  print('ログイン成功')
  #ダウンロード
  sleep(5)
  current=[]
  elements=driver.find_elements_by_xpath('//label[@class="linkText fileDownload"]')#elementに置き換えて最新版のみ
  for i in range(len(elements)):
    elements[i].click()
  dl_wait = True
  while dl_wait:
    sleep(5)
    dl_wait=False
    for fname in os.listdir("/content/"+str(filepath)):
            if fname.endswith('.crdownload'):
              dl_wait = True
  %cd ../
  import shutil
  shutil.make_archive(filepath, 'zip', root_dir='/content/',base_dir=str(filepath))
  sleep(1)
  from google.colab import files
  files.download("/content/"+str(filepath)+".zip")
```
